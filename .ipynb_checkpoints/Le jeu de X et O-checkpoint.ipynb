```python
#modélisation du tableau de jeu
import numpy as np
tab1=np.array([['-','-','-'],['-','-','-'],['-','-','-']])
tab=tab1.T
#les fontions pour l'affichage du tableau modifié
def utilisateur1(a):
    if a=='1':
        tab[0,0]='X'
        print(tab)
    elif a=='2':
        tab[0,1]='X'
        print(tab)
    elif a=='3':
        tab[0,2]='X'
        print(tab)
    elif a=='4':
        tab[1,0]='X'
        print(tab)
    elif a=='5':
        tab[1,1]='X'
        print(tab)
    elif a=='6':
        tab[1,2]='X'
        print(tab)
    elif a=='7':
        tab[2,0]='X'
        print(tab)
    elif a=='8':
        tab[2,1]='X'
        print(tab)
    elif a=='9':
        tab[2,2]='X'
        print(tab)
    else:
        print('choississez entre 1 et 9')
        
def utilisateur2(a):
    if a=='1':
        tab[0,0]='O'
        print(tab)
    elif a=='2':
        tab[0,1]='O'
        print(tab)
    elif a=='3':
        tab[0,2]='O'
        print(tab)
    elif a=='4':
        tab[1,0]='O'
        print(tab)
    elif a=='5':
        tab[1,1]='O'
        print(tab)
    elif a=='6':
        tab[1,2]='O'
        print(tab)
    elif a=='7':
        tab[2,0]='O'
        print(tab)
    elif a=='8':
        tab[2,1]='O'
        print(tab)
    elif a=='9':
        tab[2,2]='O'
        print(tab)
    else:
        print('choississez entre 1 et 9')
        
score1=0
score2=0
#Début du jeu
#Affichage du tableau
print (tab)
#Choix du signe uniquement par le premier joueur
signe1=input(("Le premier joueur choisis son signe (X/O) : "))

if signe1=='X':
    signe2='O'
    print('Le signe du joueur 2 est : ',signe2)
else:
    signe2='X'
    print('Le signe du joueur 2 est : ',signe2)
    
    
    
while not (score1==100) or (score2==100):
    while not tab[0,0]==tab[0,1]==tab[0,2] or tab[0,0]==tab[1,0]==tab[2,0] or tab[2,0]==tab[2,1]==tab[2,2] or tab[0,2]==tab[1,2]==tab[2,2] or tab[0,2]==tab[1,1]==tab[2,0] or tab[0,0]==tab[1,1]==tab[2,2]:
        #Choix des premiers emplacement par les deux joueurs   
        print('Le joueur 1 commence. Choississez votre emplacement grâce aux chiffres. Les cases sont numéroté de 1 à 9')
        a=input("Joueur 1, 1er Emplacement: ")
        if signe1=='X':
            utilisateur1(a)
        else:
            utilisateur2(a)
        b=input("Joueur 2, 1er Emplacement: ")
        if signe1=='X':
            utilisateur2(b)
        else:
            utilisateur1(b)
        #Choix des autres emplacements par les deux joueurs 
        # l'écriture 1er et 2ieme ne se ressemble pas c'est pourquoi j'ai séparé sa
        p=2
        for i in range(0,3):
            a=input("Joueur 1,"+ str(p)+"ième Emplacement: ")
            if signe1=='X':
                utilisateur1(a)
            else:
                utilisateur2(a)

            b=input("Joueur 2,"+ str(p)+"ième Emplacement: ")
            if signe1=='X':
                utilisateur2(b)
            else:
                utilisateur1(b)

            p=p+1
            break #au cas ou la condition du while n'est jamais rempli et que for à finit de s'implémenter
        #Vérifier qui gagne la manche 
        if tab[0,0]==tab[0,1]==tab[0,2]==signe1 or tab[0,0]==tab[1,0]==tab[2,0]==signe1 or tab[2,0]==tab[2,1]==tab[2,2]==signe1 or tab[0,2]==tab[1,2]==tab[2,2]==signe1 or tab[0,2]==tab[1,1]==tab[2,0]==signe1 or tab[0,0]==tab[1,1]==tab[2,2]==signe1:
            print("C'est le joueur 1 qui gagne cette manche")
            score1=+10
        elif (tab[0,0]==tab[0,1]==tab[0,2]==signe2) or (tab[0,0]==tab[1,0]==tab[2,0]==signe2) or (tab[2,0]==tab[2,1]==tab[2,2]==signe2) or (tab[0,2]==tab[1,2]==tab[2,2]==signe2) or (tab[0,2]==tab[1,1]==tab[2,0]==signe2) or (tab[0,0]==tab[1,1]==tab[2,2]==signe2):
            print("C'est le joueur 2 qui gagne cette manche")
            score2=+10
        else:
            print("Personne ne gagne cette manche rejouer")

        tab1=np.array([['-','-','-'],['-','-','-'],['-','-','-']])
        tab=tab1.T

# Finir le jeu
if score1==100:
    print('Félicitation au joueur 1. Il remporte ce jeu à',score1,'contre',score2)
if score2==100:
    print('Félicitation au joueur 2. Il remporte ce jeu à',score2,'contre',score1)
```
